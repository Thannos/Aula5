# Arredondamento e truncamento

```python
#truncamento e arredondamento  menor valor deve ser |0.1x10^-3| 
def justdoit(x,count,limite):
  if(x==0.0):
    #print("0.1 * (10 ^ (-3))")
    return 0.1, -3
  else:
      if(count>=limite): 
        a=x-int(x) 
        x=int(x)
        if(a>=0.5):
          x+=1

        cont=0
        while(x>1):
          x/=10
          cont+=1 
            

        #print(round(x,count)," * (10 ^",cont-count,")")
        p=round(x,count)
        return p, (cont-count)
      else:
        cont=0
        while(x>=1):
          x=x/10
          cont+=1

        #print(round(x,cont)," * (10 ^",cont-count,")")
        p=round(x,count)
        return p, (cont-count)
    

def especial(num, exp):
    return (num/10), (exp-1)
```

# Cálculo dos Erros 

```python
#absoluto
def sabsoluto(x,y): # Absoluto de x e Absoluto de y
    return x+y

#relativo
def srelativo(x,y,Ax,Ay): #x relativo, y relativo, Absoluto de x e Absoluto de y
    return (sabsoluto(Ax,Ay))/(x+y)

#absoluto
def mabsoluto(x,y,ax,ay): # relativo de x, relativo de y, Absoluto de x e Absoluto de y
    return (x*ay)+(y*ax)

#relativo
def mrelativo(x,y,Ax,Ay): #x relativo, y relativo, Absoluto de x e Absoluto de y
    return (mabsoluto(x,y,Ax,Ay))/(x*y)

#absoluto
def babsoluto(x,y): # Absoluto de x e Absoluto de y
    return x-y

#relativo
def brelativo(x,y,Ax,Ay): #x relativo, y relativo, Absoluto de x e Absoluto de y
    return (babsoluto(Ax,Ay))/(x-y)
```

```python
limite=3
from random import uniform
i=0
while(i<3):
    i+=1
    print("================================================\ninteração ",i,"\n================================================")
    print("Criando um valor para x")
    x=uniform(0,99.99)

    print("Criando um valor para y")
    y=uniform(0,99.99)

    print("Criando um valor para z")
    z=uniform(0,99.99)

    print("Criando um valor para t")
    t=uniform(0,99.99)

    print("\n\n\n")
    x1=3
    count=0
    h=0
    while(h!=x1):
      x*=10
      count+=1
      h+=1
    numx,expx=justdoit(x,count,limite)
    print("número x = ",(x/10**count),"\nnúmero x transformado = ",(numx*10**(count-1)),"\nErro Absoluto = ",(x/10**count)-(numx*10**(count-1)))
    #print("expoente x = ",expx)
    EAx=(x/10**count)-(numx*10**(count-1))
    x=numx
    print("\n\n\n")
    y1=3
    count=0
    h=0
    while(h!=y1):
      y*=10
      count+=1
      h+=1
    numy,expy=justdoit(y,count,limite)
    print("número y = ",(y/10**count),"\nnúmero y transformado = ",(numy*10**(count-1)),"\nErro Absoluto = ",(y/10**count)-(numy*10**(count-1)))
    #print("expoente y = ",expy)
    EAy=(y/10**count)-(numy*10**(count-1))
    y=numy
    print("\n\n\n")
    z1=3
    count=0
    h=0
    while(h!=z1):
      z*=10
      count+=1
      h+=1
    numz,expz=justdoit(z,count,limite)
    print("número z = ",(z/10**count),"\nnúmero z transformado = ",(numz*10**(count-1)),"\nErro Absoluto = ",(z/10**count)-(numz*10**(count-1)))
    #print("expoente z = ",expz)
    EAz=(z/10**count)-(numz*10**(count-1))
    z=numz
    print("\n\n\n")
    t1=3
    count=0
    h=0
    while(h!=t1):
      t*=10
      count+=1
      h+=1
    numt,expt=justdoit(t,count,limite)
    print("número t = ",(t/10**count),"\nnúmero t transformado = ",(numt*10**(count-1)),"\nErro Absoluto = ",(t/10**count)-(numt*10**(count-1)))
    #print("expoente t = ",expt)
    EAt=(t/10**count)-(numt*10**(count-1))
    t=numt

    print("\n\n\n\n")
    #(x+y)z-t
    def soma(numx, numy, expx, expy):
        if (expx == expy):
            n1=numx+numy
            return n1,expx
        else:
            if(expx<expy):
                numx/=10
                expx+=1
                soma(numx, numy, expx, expy)
            elif(expx>expy):
                numy/=10
                expy+=1
                soma(numx, numy, expx, expy)

    def multiplica(n1, numz, expx, expz): 
        n1*=numz
        expx*=expz
        return n1,expx

    def subtrai(n2, numt, exp2, expt):
        if(exp2==expt):
            n3=n2-numt
            return n3,expt
        else:
            if(exp2 > 3):
                n2=n2*10**3
                n2=round(n2,3)
                exp2=3
                subtrai(n2, numt, exp2, expt)
            else:
                exp2+=1
                n2/=10
                subtrai(n2, numt, exp2, expt)



    if(EAx<0):
        EAx*=-1
    if(EAy<0):
        EAy*=-1
    EAs=srelativo(numx,numy,EAx,EAy) #x relativo, y relativo, Absoluto de x e Absoluto de y
    print(numx,end=" ")
    numx,expx=soma(numx, numy, expx, expy)
    if((expx>3) or (numx>1)):
        numx,expx=especial(numx,expx)
    if((numx-round(numx,3))<0.0005):
        numx=round(numx,3)
    else:
        numx=round(numx,3)+1
    print("+",numy,"=",numx,"* 10 **",expx)
    print("RA da soma =",EAs)



    if(EAz<0):
        EAz*=-1
    EAm=mrelativo(numx,numz,EAs,EAz) #x relativo, y relativo, Absoluto de x e Absoluto de y
    print(numz,end=" ")
    numz,expz=multiplica(numx, numz, expx, expz)
    if((numz-round(numz,3))<0.0005):
        numz=round(numz,3)
    else:
        numz=round(numz,3) + 0.001
    if((expz>3) or (numz>1)):
        numz,expz=especial(numz,expz)
    print("*",numx,"=",numz," * 10 **",expz)
    print("RA da multiplicacao = ",EAm)



    if(EAt<0):
        EAt*=-1
    EAtotal=srelativo(numz,numt,EAm,EAt) #x relativo, y relativo, Absoluto de x e Absoluto de y
    print(numt,end=" ")
    numt,expt=subtrai(numz, numt, expz, expt)
    if((numt-round(numt,3))<0.0005):
        numt=round(numt,3)
    else:
        numt=round(numt,3) + 0.001
    if((expt>3) or (numt>1)):
        numt,expt=especial(numt,expt)
    print("-",numz,"=",numt," * 10 **",expt)
    print("RA da Subtração = ",EAtotal)

    print("\n\n\n\n")

    print("(",x,"+",y,") * ",z,"-",t," = ",numt," * 10 **",expt)
    print("Erro Relativo da opreção = ",EAtotal)
```
